```python
df.drop(df.columns[0], axis=1, inplace=True)
```

```python
def label_bid (row):
  brands = ['Apple','Samsung','Xiaomi','Huawei','Sony','Nokia',
            'Oppo','Vivo','LG','OnePlus','Realme','Infinix','Google',
            'ZTE','Asus','HTC','E-tel','Greentel','Lenovo','Dialog',
            'Sony Ericsson','Moto','Acer','BlackBerry','China Mobile',
            'Micromax','Ag-tel','Alcatel','iPro','Zigo','Microsoft']
  for id,brand in enumerate(brands):
    if row['Brand'] == brand:
      return id
  return -1

def label_cdn (row):
  if row['Condition'] == 'New':
    return 0
  if row['Condition'] == 'Used':
    return 1
  return -1

  ##########

def ds_BrandFilter(dataframe,brands):
  frames = ['']*len(brands)
  for num, brand in enumerate(brands):
    new_frame = dataframe[(dataframe["Brand"] == brand)]
    frames[num] = new_frame
  return pd.concat(frames)

def ds_Condition(dataframe,condition):
  #filterd_df = df[(df["Condition"] == condition)]
  return dataframe[(dataframe["Condition"] == condition)]
```

```python
df.insert(3,'Brand_id',0)
df.insert(5,'Condition_id',0)
```

```python
df.shape
```

```python
df['Brand_id'] = df.apply (lambda row: label_bid(row), axis=1)
df['Condition_id'] = df.apply (lambda row: label_cdn(row), axis=1)
```

```python
filterd_df = ds_Condition(df,'New')
filterd_df = ds_BrandFilter(filterd_df,['Apple','Samsung'])
```
